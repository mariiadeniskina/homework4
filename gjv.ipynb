```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.interpolate as sc
from scipy.interpolate import lagrange as lagrange
import numpy.polynomial.polynomial as npp
```

```python
g = np.load('chudo.npz')
```

```python
g.files
```

```python
x = g['arr_0']
y = g['arr_1']
```

```python
plt.plot(x, y)
plt.show()
```

```python
sad = lagrange(x, y)
```

```python
ness = np.arange(-5, 5,0.0001)
```

```python
plt.plot(ness, sad(ness))
plt.show()
```

```python
npp.Polynomial(sad).coef
```

```python
# Все коэффициенты примерно 0, кроме того, который при x^3
```

```python
r = [_**3 for _ in ness]
plt.plot(ness, r)
plt.plot(x, y)
plt.show()
```

```python
# ГРафики примерно совпадают, значит, искомая функция x^3
```
