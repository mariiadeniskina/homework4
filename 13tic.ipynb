```python
import numpy as np
import matplotlib.pyplot as plt
def f(xk,yk,xnew):
    sum = 0
    for k in range(len(yk)):
        p1=1
        p2=1
        for i in range(len(xk)):
            if k == i:
                p1=p1*1
                p2=p2*1
            else:
                p1=p1*(xnew-xk[i])
                p2=p2*(xk[k]-xk[i])
        sum += yk[k]*p1/p2
    return sum    
```

```python
def w(x):
    return 1/(25*x**2 + 1)
```

```python
xbu = np.arange(-1, 1.1, 0.2)
ybu = [w(_) for _ in xbu]
```

```python
xcheb = [np.cos(np.pi*(2*k)/20) for k in range(1,11)]
ycheb = [w(_) for _ in xcheb]
```

```python
xfluf = np.arange(-1, 1, 0.0001)
y1 = [f(xbu,ybu,_) for _ in xfluf]
y2 = [f(xcheb,ycheb,_) for _ in xfluf]
```

```python
plt.plot(xfluf,y1)
plt.plot(xfluf,y2)
plt.show()
```
